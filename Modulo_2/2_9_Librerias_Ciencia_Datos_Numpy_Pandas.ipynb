## Librerías Python para Ciencia de Datos

### Numpy
Los arrays (n-dimensionales) de NumPy son más rápidos y compactos que las listas de Python. Un array consume menos memoria y es conveniente de utilizar. NumPy utiliza mucho menos memoria para almacenar datos y proporciona un mecanismo para especificar los tipos de datos. Esto permite optimizar aún más el código.

* Más información sobre Numpy en [Numpy fundamentals](https://numpy.org/doc/stable/user/basics.html)

### Iniciación y operaciones con arrays:


```python
# Creamos una lista Python con los números del 0 al 4
l = list(range(5))
print("Lista l:", l)

# Creamos un array NumPy con los números del 0 al 4
a = np.arange(5)
print("Array a: ", a)

print("type(l):", type(l))  # Mostramos el tipo de dato de 'l'
print("type(a):", type(a))  # Mostramos el tipo de dato de 'a'
```

```python
# Creamos un array NumPy con números del 1 al 20, con paso 3
a = np.arange(1, 21, 3)
print(f"\n- a = {a}")

# Creamos un array NumPy con números impares del 1 al 19
a = np.arange(1, 20, 2)
print(f"\n- a = {a}")

# Mostramos el tipo de dato del array 'a'
print("\n- Tipo de datos de a:", type(a))
```

```python
# Creamos una nueva lista Python
l = [1, 2, 4, 6]
print("Lista l=", l)
print("tipo(l):", type(l))

# Convertimos la lista Python en un array NumPy
alist = np.array(l)
print("\n- alist:", alist)
print(" = Tipo de datos de alist:", type(alist))

# Mostramos la dimensionalidad (shape) del array NumPy
print("Dimensiones:", alist.shape)
```

```python
console.rule("[cyan]Creación de arrays[/cyan]")

# Array unidimensional
print("\n=== Array unidimensional ===")
x = np.linspace(0, 9, 21)


# Array unidimensional con elementos específicos
print("=== Array unidimensional con elementos específicos ===")
y = np.array([1, 3, 5, 7, 9])
print(f"y = {y}\n")

# Array unidimensional con elementos aleatorios
print("=== Array con elementos aleatorios ===")
z = np.random.rand(3)
print(f"z = \n{z}\n")
z = np.random.rand(1)
print(f"z = \n{z}\n")

# Generamos números aleatorios y calculamos la media y desviación estándar
a = np.random.rand(100000000)
print("-a shape:", a.shape)
print("media:", np.mean(a))
print("std:", np.std(a))
print("pct:", np.percentile(a, 25))
```

```python
# Array multidimensionales con elementos aleatorios
print("=== Array multidimensionales con elementos aleatorios ===")
z = np.random.rand(4, 3)  # Crea un array de 4 filas y 3 columnas con números aleatorios entre 0 y 1
print(f"z = \n{z}")
print(f"Forma de z: {z.shape}")  # Muestra las dimensiones del array
print(f"Tipo de datos de z: {z.dtype}\n")  # Muestra el tipo de datos del array
z = np.rint(z).astype(np.int8)
print("Int8:", z)
print(f"Tipo de datos de z: {z.dtype}\n")  # Muestra el tipo de datos del array

# Array de ceros
print("=== Array de ceros ===")
zeros = np.zeros((2, 5))
print(f"zeros = \n{zeros}\n")

# Array de unos
print("=== Array de unos ===")
ones = np.ones((3, 3))
print(f"ones = \n{ones}\n")
```

### Manipulación de arrays multidimensionales en Python
* Creación de arrays 2D, conversión de tipos de datos (de float a int), redondeo, y operaciones con arrays como la identificación de valores únicos.
* Trabajar con arrays de elementos mixtos y generar arrays aleatorios de gran tamaño.

La utilidad de estas operaciones radica en su eficiencia para manejar grandes conjuntos de datos numéricos, esenciales en análisis estadístico, procesamiento de imágenes, y machine learning.

```python
console.rule("[cyan]Arrays de dimensión superior[/cyan]")

# Array de dimensión superior
print("\n=== Array de dimensión superior ===")
y = np.array([[1.48889, 2.89, 3.777], [4.324353463, 5, 6], [1, 2, 5], [2, 2, 2]], np.float16)
print(f"y = {y}\nShape: {y.shape}")
y = np.rint(y).astype(np.int8) # El método astype nos permite modifica el tipo de dator del array
print(f"\nConversión a enteros más próximos => y: \n{y}")

# Array bidimensional con elementos mixtos
print("\n=== Array bidimensional con elementos mixtos ===")
a = np.array([[1, 2, 2, 4, 3, 4], [1, 22.345245, 2, 42, 3, 4]])
print("Array a:", a)

# Obtener los valores no repetido
z, c = np.unique(a, return_counts=True)
print(f"= Valores únicos en el array: {z}")
print("Cuántas veces se repite cada valor:", c)

# Contar valore únicos
z = np.random.rand(5040, 40)
print("Valores únicos en z:", len(np.unique(z)))
print("Tamaño; 5040*40=", 5040*40)
```

### Estadística descriptiva básica

Métodos estadísticos básicos que se pueden aplicar a arrays NumPy, esenciales para el análisis exploratorio de datos en ciencia de datos: calcular el valor máximo, mínimo, media, desviación estándar y percentiles de un array.

```python
console.rule("[cyan]Métodos varios a ejecutar sobre arrays[/cyan]")

y=z
print("Y:", y)

print("\nValor máximo:")
print("Max:", np.max(y))

print("\nValor mínimo:")
print("Min:", np.min(y))

print("\nMedia:")
print("Media:", np.mean(y))

print("\nDesviación estándar:")
print("Std:", np.std(y))

print("\nPercentil 25:")
print("Pct:", np.percentile(y, q=25))
print("Pct Q99:", np.percentile(y, q=99))
```

```python
# Complemento de estadística descriptiva

print("\nMediana:")
print("Mediana:", np.median(y))

print("\nVarianza:")
print("Var:", np.var(y))

print("\nSuma de todos los elementos:")
print("Suma:", np.sum(y))

print("\nProducto de todos los elementos:")
print("Producto:", np.prod(y))

print("\nRango (diferencia entre máximo y mínimo):")
print("Rango:", np.ptp(y))

print("\nCuartiles:")
print("Q1, Q2, Q3:", np.percentile(y, [25, 50, 75]))
```

```python
console.rule("[cyan]Slicing de arrays[/cyan]")
# Slicing con NumPy arrays

c = np.arange(1, 11, 1)
print("c:", c)

print("\nPrimeros 2 valores:")
print(c[:2])

print("\nÚltimos 8 elementos:")
print(c[2:])

print("\nÚltimos 2 valores:")
print(c[-2:])

print("\Primeros 8 valores:")
print(c[:-2])
```

## Numpy filtros

```python
console.rule("[cyan]Filtros sobre arrays (datos de imagen)[/cyan]")

import matplotlib.pyplot as plt
import numpy as np
from skimage import data, filters

# Cargar la imagen de muestra
image = data.camera()
print("image shape:", image.shape)

# Crear una figura con subplots
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

# Mostrar la imagen original
ax[0].imshow(image)
ax[0].set_title('Imagen original')

# Aplicar filtro de desenfoque Gaussiano
blurred = filters.gaussian(image, sigma=16)
ax[1].imshow(blurred, cmap='gray')
ax[1].set_title('Filtro de desenfoque Gaussiano')

# Aplicar filtro Sobel para detectar bordes
edges = filters.sobel(image)
ax[2].imshow(edges, cmap='gray')
ax[2].set_title('Filtro Sobel para detección de bordes')

# Ajustar los espacios entre subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, filters

# Cargar la imagen de muestra
image = data.camera()

# Calcular los percentiles 5 y 95 de la imagen
percentile_50 = np.percentile(image, 50)
percentile_99 = np.percentile(image, 99)

# Aplicar umbral a la imagen para eliminar valores por debajo del percentil 5
# y por encima del percentil 95
thresholded = np.clip(image, percentile_50, percentile_99)

# Crear una figura con subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Mostrar la imagen original
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Imagen original')

# Mostrar la imagen con valores fuera del rango recortados
ax[1].imshow(thresholded, cmap='gray')
ax[1].set_title('Valores recortados (5%-95%)')

plt.show()

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].hist(image)
ax[1].hist(thresholded)
```

```python
console.rule("[cyan]Otras operaciones básicas[/cyan]")

# Crear un DataFrame desde un diccionario
data = {
    'Nombre': ['Juan', 'María', 'Pedro', 'Ana'],
    'Edad': [25, 30, 40, 35],
    'Ciudad': ['Madrid', 'Madrid', 'Valencia', 'Sevilla'],
    'Salario': [255000, 24444, 30000, 42222]
}
df = pd.DataFrame(data)
print("Datos:\n", df)


# Mostrar las últimas filas del DataFrame
print("Últimas filas:\n", df.tail(2))

# Filtrar el DataFrame basado en una condición
filtered_df = df[df['Edad'] > 25]
print("\nFiltrado por edad:\n", filtered_df)

# Ordenar el DataFrame por una columna
sorted_df = df.sort_values(['Edad'], ascending=True)
print("\nOrdenado por edad:\n", sorted_df)
```

```python
console.rule("[cyan]Operaciones básicas con DataFrames de Pandas[/cyan]")

# Seleccionar columnas específicas
selected_columns = df[['Nombre', 'Salario']]
print("\n\nSelección de columnas 'Nombre' y 'Salario':")
print(selected_columns)

# Agregar una nueva columna
df['Años_Experiencia'] = [3, None, 10, None]
print("\nDataFrame con nueva columna 'Años_Experiencia':")
print(df)
```

```python
print(df.isnull().sum())
```

```python
# Renombramos la columna Salario a Ingresos
df = df.rename(columns={'Salario': 'Ingresos'})
print(df)

# Calcular estadísticas por grupo
group_stats = df.groupby('Ciudad')['Ingresos'].mean()
print("\n\nSalario promedio por ciudad:")
print(group_stats)

# Renombrar columnas
df = df.rename(columns={'Años_Experiencia': 'Experiencia', 'Ingresos': 'Salario'})
print("\n\nDataFrame con columna renombrada:")
print(df)

# Agregar una nueva columna calculada al DataFrame
df['Edad_Doble'] = df['Edad'] * 2
print("\nDataFrame con columna adicional:\n", df)
```

```python
# Contar los valores únicos en una columna.
counts = df['Ciudad'].value_counts()
print("\nValores únicos:\n", counts)

# Calcular la media de una columna
mean_age = df['Edad'].mean()
print("\nMedia de edad:", mean_age)

# Eliminar una columna del DataFrame
console.rule("Eliminar una columna del DataFrame")
df = df.drop('Edad_Doble', axis=1)
print("\nDataFrame sin la columna Edad_Doble:\n", df)
```

# Seaborn

Seaborn es una librería de visualización de datos basada en Matplotlib que proporciona una interfaz de alto nivel para crear gráficos atractivos y informativos. Está especialmente diseñada para trabajar con DataFrames de pandas y se integra bien con las funcionalidades de análisis de datos de Python.

Enlace a la documentación oficial de Seaborn: [Documentación de Seaborn](https://seaborn.pydata.org/)



```python
console.rule("[cyan]Gráficos con Seaborn[/cyan]")

import seaborn as sns

# Cargar el conjunto de datos "iris"
iris = sns.load_dataset('iris')
print("IRIS dataset:\n", iris)
print("IRIS tipo:", iris.columns)
print("Variedades:", np.unique(iris['species'].values))
```

```python
# Gráfico de distribución (histograma y KDE)
sns.histplot(iris['sepal_width'], bins=10)
import matplotlib.pyplot as plt

mean = iris['sepal_length'].mean()
std = iris['sepal_length'].std()
plt.title(f'Distribución de la longitud del sépalo: Media = {round(mean,2)} Std = {round(std,2)}')
plt.xlabel('Longitud del sépalo')
plt.ylabel('Cuentas')
plt.show()
```

```python
# Gráfico de dispersión (scatter plot)
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Relación entre la longitud y el ancho del sépalo')
plt.xlabel('Longitud del sépalo')
plt.ylabel('Ancho del sépalo')
plt.show()
```

```python
# Gráfico de barras (bar plot)
sns.barplot(x='species', y='petal_length', data=iris, color="cyan")
plt.title('Longitud promedio del pétalo por especie')
plt.xlabel('Especie')
plt.ylabel('Longitud promedio del pétalo')
plt.show()
```

```python
# Gráfico de cajas (box plot)
sns.boxplot(data=iris)
plt.title('Distribución del ancho del pétalo')
plt.xlabel('Variable')
plt.ylabel('Longitud/Anchura [cm]')
plt.show()
```
